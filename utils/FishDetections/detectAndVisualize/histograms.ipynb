```python
import json
# Load the data from the extracted JSON file
json_file_path = '/Users/apaula/src/SharkVideoAnalysis/frame_0_properties.json'

with open(json_file_path, 'r') as file:
    data = json.load(file)
```

```python
import matplotlib.pyplot as plt

# Extracting the values for density, polarization, and mean orientation
density = [item['density'] for item in data]
polarization = [item['polarization'] for item in data]
mean_orientation = [item['mean_orientation'] for item in data]

# Plotting histograms for each property
plt.figure(figsize=(15, 5))

# Density Histogram
plt.subplot(1, 3, 1)
plt.hist(density, bins=30, color='blue', alpha=0.7)
plt.title('Density Histogram')
plt.xlabel('Density')
plt.ylabel('Frequency')

# Polarization Histogram
plt.subplot(1, 3, 2)
plt.hist(polarization, bins=30, color='green', alpha=0.7)
plt.title('Polarization Histogram')
plt.xlabel('Polarization')

# Mean Orientation Histogram
plt.subplot(1, 3, 3)
plt.hist(mean_orientation, bins=30, color='red', alpha=0.7)
plt.title('Mean Orientation Histogram')
plt.xlabel('Mean Orientation')

plt.tight_layout()
plt.show()
```

```python
# Extracting x and y values from the data
x_values = [item['x'] for item in data]
y_values = [item['y'] for item in data]

# Plotting histograms for x and y values
plt.figure(figsize=(12, 5))

# X Values Histogram
plt.subplot(1, 2, 1)
plt.hist(x_values, bins=30, color='purple', alpha=0.7)
plt.title('X Values Histogram')
plt.xlabel('X Value')
plt.ylabel('Frequency')

# Y Values Histogram
plt.subplot(1, 2, 2)
plt.hist(y_values, bins=30, color='orange', alpha=0.7)
plt.title('Y Values Histogram')
plt.xlabel('Y Value')

plt.tight_layout()
plt.show()
```
